Matplotlib comes with a wide variety of plots. Plots help us to understand the trends,patterns and to make correlations. They are typically instruments for reasoning about qualitative information

```python
#line charts are used to find the relation b/w the two variables;
#uptrend and downtrend
```

```python
#Simple Example

#x-axis values
x=[50000,200000,900000,400000,700000]

#y-axis values
y=[2003,2004,2005,2006,2007]

#function to plot
plt.plot(y,x,color='r')

#function to show the plot
plt.show()
```

```python
#Bar chart is used to compare the data 

#x-axis values
x=[50000,200000,900000,400000,700000]

#y-axis values
y=[2003,2004,2005,2006,2007]

#function to plot
plt.bar(y,x,color='green')

#function to show the plot
plt.show()
```

```python
#Bar plot

#x-axis values
x=['Python','ML','Full Stack','AWS','AZURE']

#y-axis values
y=[10,5,8,4,2]

#function to plot
plt.bar(x,y,color='purple')

#function to show the plot
plt.show()
```

```python
#horizontal bar plot

x_values=[5,6,3,7,2]
y_values=["A","B","C","D","E"]

#Adding on "h" after bar will flip the graph
plt.barh(y_values,x_values,color="yellowgreen")

#function to show the plot
plt.show()
```

```python
#scatter plot is used for correlation b/w two variables
#range is (-1 to 1)
#positive and negative correlation
#(-1 to 0) negative correlation
#(0 to 1) positive correlation
#it is the visual represntation to find out relation b/w variables
```

```python
#pie chart is used see the proportion of each category in a dataset
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,29,12]

plt.pie(students,labels=langs,autopct='%.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
my_explode=(0,0.1,0.4,0,0)
langs=['C','C++','Java','Python','PHP']
students=[23,17,35,29,12]

plt.pie(students,labels=langs,shadow=True, autopct='%.1f%%',explode=my_explode)
plt.title('Pie Chart')
plt.show()
```

```python
#Voilin plots
#use to visualize the distribution of data.
#it is used  for the numerical data and categoriacal data.
#it is the hybrid of a box plot and a kernel density plot, which shows peaks in the data. 
```

Voilin plots are similar to box plots, except that they also show the probability of the data at different values.

```python
collectn_1=np.random.normal(100,10,200) #(mean,std_dev,no.of random numbers)
collectn_2=np.random.normal(80,30,200)
collectn_3=np.random.normal(90,20,200)
collectn_4=np.random.normal(70,20,200)

#combine these different collections into a list
data_to_plot=[collectn_1,collectn_2,collectn_3,collectn_4]
```

```python
plt.violinplot(data_to_plot)
plt.show()
```
